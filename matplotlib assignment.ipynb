```python
titanic = pd.read_csv('https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv', skip_blank_lines=True)
```

```python
titanic
```

```python
titanic.drop(titanic.tail(1).index,inplace=True)
```

```python
fare_mode = titanic.fare.mode()
fare_mode
```

```python
titanic.fare.fillna(fare_mode, inplace=True)
```

```python
titanic.isna().transpose()
```

1. Create a pie chart presenting the male/female proportion

```python
count = titanic['sex'].value_counts()
```

```python
fig1, (ax1) = plt.subplots()
ax1.pie(count.values.tolist(), labels=count.index.values.tolist(), autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
titanic = titanic.dropna(subset=['sex'])

mapping = {'male' : 'blue', 'female' : 'red'}
```

```python
fig2, (ax2) = plt.subplots()

ax2.scatter(titanic['age'], titanic['fare'], alpha=0.5, c=titanic['sex'].map(mapping), label=mapping)
ax2.set_xlabel('Fare')
ax2.set_ylabel('Age')
ax2.set_title('Fare paid and the Age')
plt.legend()
plt.show()
```
